```python
#The following set of code uses the csv file with the type or genre of music that people of certain age prefeer to listen.
#The target or the prediction is about the kind of music -  given the age and gender
#We are using DecisionTreeClassifier for creating the model and predicting the outcome. 
#y which is the prediction is determined by reading just the genre column of music_data dataframe. 
#X is the set of features used to make the prediction.
# inroder to make a prediction we will have to divide the data into training set and testing set or validation set , 
#done using train_test-split library
# Then the model is trained using command fit using training data.
# Thereafter prediction is made using validation data.
#The accusracy of prediction in this case is determined using accuracy_score between what we predict and valid_y. 
# The library used for this is from scikit learn. 

# Not always do we keep doing teh steps of data split >model fit and predict.
# The alternative is that we use joblib library to persist the prediction. Using joblib.dump command we push the prediction to a 
#file : example: mosic-recomender.joblib
# Using thi joblib prediction we can make predictions by loading the joblib model file.
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib
from sklearn import tree

music_data = pd.read_csv('music.csv')
# music_data
model = DecisionTreeClassifier()
X = music_data.drop(columns = ['genre'])
y = music_data['genre']
train_X, valid_X, train_y, valid_y = train_test_split(X,y, test_size=.2, random_state=0)
model.fit(train_X,train_y)   
```

```python
# make prediction
# predict = model.predict(valid_X)
#Determine the accuracy of prediction 
# print(accuracy_score(valid_y, predict))
```

```python
tree.export_graphviz(model, out_file = 'music-recommender.dot',
                    feature_names= ['age','gender'],
                    class_names = sorted(y.unique()),
                                       label='all',
                                       rounded = True,
                                       filled = True)
```
